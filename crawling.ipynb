```python
def replace_df(df):
    mapping = {'-': 0} 
    
    replace_dict = {}
    
    for colum in df.columns:
        replace_dict[colum] = mapping
    return df.replace(replace_dict)
```

```python
def crawling_league_teams(team_id, api_delay_term=5):

    """
    cawling league team_id and team name datas
    
    parameter ---------------------------------------------------------------
    team_id : one of you want league team_id & parameter data type int or str
    
    return ------------------------------------------------------------------
    crawling league team_id, team_name datas belong team_id parameter
    return pandas dataframe columns=team_id, team_name
    
    """
    
    # connect webdriver
    url = "https://www.whoscored.com/Teams/" + str(team_id)
    driver = webdriver.Chrome('C:/Users/jaemini_man/Desktop/soccer/chromedriver')
    driver.get(url)

    # wait get league team datas
    time.sleep(api_delay_term) 
    
    # make pandas dataframe
    team_df = pd.DataFrame(columns=["team_id","team_name"])
    
    # get team datas
    teams = driver.find_elements_by_css_selector("#teams option")
    for team in teams:
        team_name = team.text
        team_id = team.get_attribute("value").split("/")[2]
        team_df.loc[len(team_df)] = {"team_id":team_id, "team_name":team_name }
        
    # close webdriver
    driver.close()
    
    return replace_df(team_df)
```

```python
def crawling_team_fixture(team_id, api_delay_term=5, seasons='2020-2021'):
        # connect webdriver
    url = "https://www.whoscored.com/Teams/" + str(team_id) +'/Fixtures'
    driver = webdriver.Chrome('C:/Users/jaemini_man/Desktop/soccer/chromedriver')
    driver.get(url)

    # wait get league team datas
    time.sleep(api_delay_term) 
    fixture_results_df = pd.DataFrame(columns=['match_up','competition','score', 'date'])
    
    elements = driver.find_elements_by_css_selector('a.horiz-match-link.result-1')
    date_elements = driver.find_elements_by_css_selector('div.col12-lg-1.col12-m-1.col12-s-0.col12-xs-0.date.fourth-col-date.divtable-data')
    for i in range(len(elements)):
        temp = elements[i].get_attribute('href').split('/')[-1]
        if seasons in temp:
            fixture_dict = {
                'match_up' : temp.split(seasons)[1][1:],
                'competition' :  ''.join(c for c in temp.split(seasons)[0] if c.isupper()),
                'score' : elements[i].text,
                'date' : date_elements[i].text
            }
            fixture_results_df.loc[len(fixture_results_df)] = fixture_dict
    
    driver.close()
    
    return fixture_results_df
```

```python
def save_team_fixture(league_df, seasons='2020-2021'):
    team_id_list = list(league_df.team_id)
    team_name_list = list(league_df.team_name)
    for i in range(len(team_id_list)):
        temp_df = crawling_team_fixture(int(team_id_list[i]), api_delay_term=4, seasons=seasons)
        temp_df.to_csv(team_name_list[i]+'_fixture_'+seasons+'.csv')
        print(team_name_list[i])
    return()
```

```python
def replace_all(df):
    sub = list()
    for i in list(df['Apps']):
        if '(' in i :
            sub.append(i.split('(')[1].split(')')[0])
        else :
            sub.append('0')
    df['sub'] = sub
    df['Apps'] = list(map(lambda x: x.split('(')[0], list(df['Apps'])))
    df['team_name'] = list(map(lambda x: x.split(',')[0], list(df['team_name'])))
    df['position'] = list(map(lambda x: x.split(' ')[1],list(df['position'])))
    return(df)          
```
