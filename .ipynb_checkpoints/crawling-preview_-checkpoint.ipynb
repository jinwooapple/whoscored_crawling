```python
def final(path,season,lis, trans=0):
    match_number = pd.read_csv(season + '_match_number.csv')['0']
    match_no = len(match_number)
    new_list = [i-trans for i in lis]
    result = crawling_seasons_add(path,season,new_list)
    results = duplicated_match_filter(season,match_no)
    return(result)
```
